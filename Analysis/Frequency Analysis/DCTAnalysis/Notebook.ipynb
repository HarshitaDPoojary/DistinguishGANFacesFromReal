```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import dct
from PIL import Image
import os

def dct2(a):
    return dct(dct(a.T, norm='ortho').T, norm='ortho')

def log_scale(coefficients, epsilon=1e-10):
    abs_coefficients = np.abs(coefficients)
    return np.log(abs_coefficients + epsilon)

def plot_dct(image_path):
    # Load image and convert to grayscale
    image = Image.open(image_path)
    image_data = np.array(image)

    # Apply 2D DCT
    dct_transformed = dct2(image_data)

    # Log scale for better visualization
    dct_log_scaled = log_scale(dct_transformed)

    # Visualize DCT
    plt.imshow(dct_log_scaled)
    plt.title(f'DCT of {os.path.basename(image_path)}')
    plt.colorbar()
    plt.show()

def process_directory(directory_path):
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        if os.path.isfile(file_path):
            plot_dct(file_path)

# Replace with your directory path
directory_path = '/content/sample_data/Dataset/real'
process_directory(directory_path)
```
